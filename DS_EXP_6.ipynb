```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart_disease.csv')
df
```

```python
### Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['resting blood pressure'], c='blue', marker='o')
plt.xlabel('age')
plt.ylabel('resting blood pressure')
plt.title('Scatter Plot')
plt.show()
```

```python
### Bar Plot
import matplotlib.pyplot as plt
plt.bar(df['age'],df['resting blood pressure'])
plt.xlabel('age')
plt.ylabel('resting blood pressure')
plt.title('Bar Plot ')
plt.show()
```

```python
### Histogram
# A histogram is a type of bar chart that represents the distribution of numerical data.
plt.hist(df['age'], bins=5, edgecolor='black')
plt.xlabel('age')
plt.ylabel('resting blood Pressure')
plt.title('Histogram')
plt.show()
```

```python
### Line Charts
plt.plot(df['age'], df['resting blood pressure'], marker='o')
plt.xlabel('age')
plt.ylabel('resting blood pressure')
plt.title('age Over Blood Pressure')
plt.grid(True)
plt.show()
```

```python
### Box Plot
sns.boxplot(x='age', y='resting blood pressure', data=df)
plt.xlabel('age')
plt.ylabel('resting blood pressure')
plt.title('Box Plot Example')
plt.show()
```
